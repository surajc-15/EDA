```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how='left')
```

```python
##to check data type
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##pie chart 
plt.pie(country_values,labels=country_names)
```

```python
##pir chart for top 3 countries
##pie chart 
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

observation:zomato maximum records or transaction from India then USA and then comes UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
#the last column had title as 0 so that is renamed 
ratings
```

1.when rating is betweeen 4.5 to 4.9  ---> Excellent
2.when rating is betweeen 4.0 to 4.4  ---> very good
3.when rating is betweeen 3.5 to 3.9  ---> good
etc


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
##plotting map with colors
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1.not rated count is very high
2.Maximum ratings are between 2.5 to 3.4

```python
##count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
